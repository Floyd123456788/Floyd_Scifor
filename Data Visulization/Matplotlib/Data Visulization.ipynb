```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

# 1Q] Bar Graph

```python
x1=[1,3,5,7,9]
y1=[5,2,7,8,2]

x2=[2,4,6,8,10]
y2=[8,2,7,8,2]

plt.bar(x1,y1,color='blue'
       )
plt.bar(x2,y2,color='green')

plt.title('bar graph example')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Horizontal Bar Graph

```python
x1=[1,3,5,7,9]
y1=[5,2,7,8,2]

x2=[2,4,6,8,10]
y2=[8,2,7,8,2]

plt.barh(x1,y1,color='red')
plt.barh(x2,y2,color='yellow')

plt.title('bar graph example')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Pie Chart 

```python
y=np.array([20,30,40,10])
labelss=['A','B','C','D']
explodes=[0.2,0,0,0]
mycolors=['red','green','orange','blue']
plt.pie(y,labels=labelss,startangle=90,explode=explodes,colors=mycolors)
plt.legend(title='Pie chart')
plt.show()
```

# Scatter Plot

```python
x=[1,5,7,8,2,3,7,6,8,9,10,17]
y=[3,4,6,2,3,7,6,3,4,5,8,9]

plt.scatter(x,y,color='blue',marker='*')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('scatter plot')
plt.show()
```

# Stacked Bar Graph

```python
X=['F','L','O','Y','D']
y1=[10,20,30,40,50]
y2=[60,70,80,90,100]

plt.bar(X,y1,color='Blue')
plt.bar(X,y2,color='Yellow',bottom=y1)
plt.legend(['y1','y2'])
plt.title('Stacked Bar')
plt.show()
```

# Multiple Line Graph

```python
x1=[1,5,7,8]
y1=[4,6,2,3]

x2=[4,9,4,1]
y2=[6,8,7,2]

plt.plot(x1,y1,color='purple',label='lineone',linewidth=5)
plt.plot(x2,y2,color='green',label='linetwo',linewidth=5)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Multiple Line Graph')
plt.grid(True)
plt.legend()
plt.show()
```
