```python
import pandas as pd
import seaborn as sns
ds=pd.read_csv("C:/Users/floyd/data/dummy_data.csv")
ds.head()
```

```python
ds.columns
```

```python
ds.info()
```

# LINE PLOT

```python
sns.set(style='dark')
sns.lineplot(x='age',y='indebt',data=ds)
```

# BAR PLOT

```python
sns.barplot(x='gender',y='isHomeOwner',data=ds,palette='plasma')
```

# BOX PLOT

```python
sns.boxplot(x='gender',y='time_spent',data=ds)
```

# Violinplot

```python
sns.violinplot(x='gender',y='time_spent',data=ds,split=True)
```

# Distplot

```python
sns.distplot(
    ds['time_spent'],bins=5,kde=True,color='red')
```
